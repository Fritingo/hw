```python
# 畫花蕊的排序 以 Fibonacci 的方式生長 

#@title Default title text { run: "auto", vertical-output: true }

plt.figure(0, figsize=(10, 10))

# 中間的花蕊數目
seed = 889  #@param {type:"slider", min:0, max:1000, step:1}


#137.51
golden_angle = 137.51 # Fibonacci 
angle = 0
for i in range(1,seed):
  angle += golden_angle 
  angle = angle%360
  if angle < 90:
    plt.scatter(i*np.cos(angle*np.pi/180),i*np.sin(angle*np.pi/180),s = 100)
  elif angle < 180:
    patch = 180 - angle
    plt.scatter(-i*np.cos(patch*np.pi/180),i*np.sin(patch*np.pi/180),s = 100)
  elif angle < 270:
    patch = angle - 180
    plt.scatter(-i*np.cos(patch*np.pi/180),-i*np.sin(patch*np.pi/180),s = 100)
  else:
    patch = 360 - angle
    plt.scatter(i*np.cos(patch*np.pi/180),-i*np.sin(patch*np.pi/180),s = 100)
  # plt.scatter(-2*np.cos((180-137.51)*np.pi/180),2*np.sin((180-137.51)*np.pi/180),s = 100),c = '#b37f00'



plt.xlim(-seed,seed)
plt.ylim(-seed,seed)
```
