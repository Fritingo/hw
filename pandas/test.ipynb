```python
import pandas as pd

# dataset from data.world (NBA Finals and MVPs)
# dataset url https://data.world/datatouille/nba-finals-and-mvps

df = pd.read_csv('./datatouille-nba-finals-and-mvps/data/data.csv')
```

```python
# check it is loaded
df.head()
```

## No 1. 如果我是湖人迷那湖人隊在1980到2010拿了幾次總冠軍？

```python
mask = (df['year'].between(1980,2010)) & (df['nba_champion'] == 'Los Angeles Lakers')
print(len(df[mask]))
df[mask]
```

## No 2. 這些年又有幾次是西區冠軍但是沒拿到總冠軍？

```python
mask = (df['year'].between(1980,2010)) & (df['nba_champion'] != 'Los Angeles Lakers') & (df['western_champion'] == 'Los Angeles Lakers')
print(len(df[mask]))
df[mask]
```

## No 3. 這些年又有哪些人是湖人隊成員拿到例行賽ＭＶＰ？

```python
mask = (df['year'].between(1980,2010)) & (df['mvp_team'] == 'Los Angeles Lakers') 
print(set(df[mask]['mvp_name']))
df[mask]
```

## No 4. 1960年到2000年例行賽MVP所屬位置的比例是多少？

```python
temp = []
for pos in set(df['mvp_position']):
    tempmask = (df['mvp_position'] == pos) & (df['year'].between(1960,2000))
    temp.append(len(df[tempmask]))
    # print(pos)
    # print(df[tempmask])
for i in range(3):
    print(list(set(df['mvp_position']))[i+1] , ': ' ,temp[i+1]/sum(temp),'%')


#df[tempmask]
```

## No 5. 2000年到2018年例行賽MVP所屬位置的比例又是多少？

```python
temp = []
for pos in set(df['mvp_position']):
    tempmask = (df['mvp_position'] == pos) & (df['year'].between(2000,2018))
    temp.append(len(df[tempmask]))
    
for i in range(3):
    print(list(set(df['mvp_position']))[i+1] , ': ' ,temp[i+1]/sum(temp),'%')
```

## No 6. 1960年到2000年例行賽MVP平均身高？

```python
mask = (df['year'].between(1960,2000))
print(sum(df[mask]['mvp_height_m'])/len(df[mask]['mvp_height_m']) ,'m')
```

## No 7. 1960年到2018年有哪些隊伍成功在冠軍賽橫掃？

```python
mask = (df['year'].between(1960,2018)) & (df['final_sweep'] == True)
df[mask]['nba_champion']
```

## No 8.例行賽MVP選手主要都是哪個國家的人？

```python
temp = []
for nationality in set(df['mvp_nationality']):
    tempmask = (df['mvp_nationality'] == nationality)
    temp.append(len(df[tempmask]))

print(list(set(df['mvp_nationality']))[temp.index(max(temp))])
```

## No 9. S.Curry在哪些年份拿到MVP?

## No 10. 1960至2018的東西區奪冠比例是？

```python
mask1 = (df['year'].between(1960,2018)) & (df['nba_champion'] == df['western_champion'])
mask2 = (df['year'].between(1960,2018)) & (df['nba_champion'] == df['eastern_champion'])
mask3 = (df['year'].between(1960,2018))
print('western_champion: ',len(df[mask1])/len(df[mask3]), '%')
print('eastern_champion: ',len(df[mask2])/len(df[mask3]), '%')
```
